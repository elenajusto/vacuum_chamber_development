```python
# Libraries
import pandas as pd
```

```python
# Get raw CSV as a dataframe
raw_df = pd.read_csv("input/example_curve.csv", skiprows=3)     # Skip OpenRocket generated headers
raw_df
```

```python
# Get altitude data only
dataStream = []     
for index, row in raw_df.iterrows():
    if str(row["Altitude (m)"]) == "nan":
        continue
    else:
        try:
            row = float(row["Altitude (m)"])
            dataStream.append(row)
        except:
            continue

print(dataStream)
```

```python
# Save altitude data into output.txt
f = open("output/output.txt", "w")

for item in dataStream:
    f.write(str(item))
    f.write(",\n")

f.close()
```
