```python
# Libraries
import pandas as pd
```

```python
# Get raw CSV as a dataframe
raw_df = pd.read_csv("input/example_curve.csv", skiprows=3)     # Skip OpenRocket generated headers
raw_df
```

```python
dataStream1 = []        # Time
dataStream2 = []        # Altitude
dataStream3 = []        # Pressure (mbar)

for item in raw_df["# Time (s)"]:
    try:
        item = float(item)
        dataStream1.append(item)
    except:
        continue

for item in raw_df["Altitude (m)"]:
    if str(item) == "nan":
        continue
    else:
        dataStream2.append(item)

for item in raw_df["Air pressure (mbar)"]:
    if str(item) == "nan":
        continue
    else:
        dataStream3.append(item)

print("Data Stream 1 - No. Items: ", len(dataStream1))
print(dataStream1)
print("Data Stream 2 - No. Items: ", len(dataStream2))
print(dataStream2)
print("Data Stream 3 - No. Items: ", len(dataStream3))
print(dataStream3)
```

```python
# Text Packet Format
# Count of Time Data
# Count of Altitude Data
# Count of Pressure Data

outputArray = []

outputArray.append(len(dataStream1))
outputArray.append(len(dataStream2))
outputArray.append(len(dataStream3))

for item in dataStream1:
    outputArray.append(item)

for item in dataStream2:
    outputArray.append(item)

for item in dataStream3:
    outputArray.append(item)

for item in outputArray:
    print(item)
```

```python
f = open("output/output.txt", "a")

for item in outputArray:
    f.write(str(item))
    f.write(",")

f.close()
```
